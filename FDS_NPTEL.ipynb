##### Dispersion 
In statistics, dispersion is the extent to which a distribution is stretched or squeezed. Common examples of measures of statistical dispersion are the variance, standard deviation, and interquartile range. <br>
###### Skewness <br>
    -Absence of symmetry
    -Extreme values in one side of a distribution
    Skewed right - Median lies between mode and mean, and mode is less than mean 
    Skewed left - Median lies between mode and mean, and is greater than mean 
    
###### Kurtosis <br>
    Peakedness of a distribution 
    -Leptokurtic : high and thin 
    -Mesokurtic : normal shape 
    -Platykurtic : flat and spread out

## Probability

- Probability is the numerical measure of the likelihood that an event will occur. <br>
- 0<= P(A) <=1 <br>
- Probabilities of all mutually exclusive and collectively exhaustive events is 1. <br>
- P(A) + P(B) + P(C) = 1 <br>

### Methods of Assigning Probabilities 
##### Classical method of assigning brobability (rules and laws) 
    -Number of outcomes leading to the event divided by the total number of outcomes possible
    -Determined a priori -- before performing the experiment
    - P(E) = n/N 
##### Relative frequency of occurence (cumulated historical data) 
    -Based on historical data 
    -Number of times the event occured divided by the number of trails
    
##### Subjective Probability (personl intution or reasoning)
    -comes from a person's intuition or reasoning 
    -based on degree of belief
