```python
import requests
from bs4 import BeautifulSoup

# url = 'http://www.whatobuy.in/old-songs-lyrics-hindi-movies/'
url = 'https://www.w3schools.com/html/'
# url = 'https://codewithharry.com/'
r = requests.get(url)
htmlparser = r.content
# print(htmlparser)
# htmlparser.prettify  ----> not a valid
```

```python
soup = BeautifulSoup(htmlparser, 'html.parser')
# soup.prettify
# soup
```

```python
anchors = soup.find_all('a')
all_links = set()

for link in anchors:
    if 'html' in link.get('href'):
        if 'https' not in link.get('href'):
            if str(link.get('href')).startswith("/html"):
                all_links.add('https://www.w3schools.com' + link.get('href'))
                print('https://www.w3schools.com' + link.get('href'))
            elif str(link.get('href')).startswith("html"):
                all_links.add('https://www.w3schools.com/html/' + link.get('href'))
                print('https://www.w3schools.com/html/' + link.get('href'))
                
print(all_links)
    
```
